```python
qid_question_dict = {}
for index, row in train_data.iterrows():
    qid_question_dict[row['qid1']] = row['question1']
    qid_question_dict[row['qid2']] = row['question2']
```

```python
len(qid_question_dict)
```

```python
duplicate_questions = train_data[train_data.is_duplicate == 1]
```

```python
duplicate_questions
```

```python
f = dict.fromkeys(qid_question_dict)
rank  = dict.fromkeys(qid_question_dict)
for x in f:
    f[x] = x
    rank[x] = 0
```

```python
def find(x):
    if f[x] != x:
        f[x] = find(f[x])
    return f[x]

def make(x, y):
    x = find(x)
    y = find(y)
    if x == y: return
    if rank[x] > rank[y]:
        f[y] = x
    else:
        if rank[x] == rank[y]:
            rank[y] += 1
        f[x] = y
```

```python
for index, row in duplicate_questions.iterrows():
    make(row['qid1'], row['qid2'])
```

```python
for x in f:
    find(x)
```

```python
f
```

```python
duplicates = list(f.items())
```

```python
duplicates = sorted(duplicates, key=lambda x:x[1])
```

```python
duplicates_set = set()
for d in duplicates:
    duplicates_set.add(str(d[0]) + ' ' + str(d[1]))
```

```python
duplicates_set
```

```python
duplicates
```

```python
duplicates_dict = dict.fromkeys(f)
for d in duplicates_dict:
    duplicates_dict[d] = set()
```

```python
for d in duplicates:
    duplicates_dict[d[1]].add(d[0])
```

```python
duplicates_dict
```

```python
duplicates_dict_reduce = dict()
for d in duplicates_dict:
    if len(duplicates_dict[d]) > 2:
        duplicates_dict_reduce[d] = duplicates_dict[d]
```

```python
sum = 0
for d in duplicates_dict_reduce:
    temp = len(duplicates_dict_reduce[d])
    sum += temp * (temp - 1)
sum
```

```python
sum = 0
new_duplicates = []
for d in duplicates_dict_reduce:
    temp = list(duplicates_dict_reduce[d])
    for a in range(len(temp)):
        x = temp[a]
        for b in range(a + 1, len(temp)):
            y = temp[b]
            z1 = str(x) + ' ' + str(y)
            z2 = str(y) + ' ' + str(x)
            if z1 not in duplicates_set and z2 not in duplicates_set:
                new_duplicates.append((x, y))
                sum += 1
sum
```

```python
new_duplicates_text = [(qid_question_dict[x[0]], qid_question_dict[x[1]]) for x in new_duplicates]
```

```python
new_duplicates_list = []
for n in new_duplicates:
    new_duplicates_list.append([-1, n[0], n[1], qid_question_dict[n[0]], qid_question_dict[n[1]], 1])
```

```python
duplicate_df = pd.DataFrame(new_duplicates_list, columns=train_data.columns)
```
